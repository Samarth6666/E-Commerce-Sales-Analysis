# KPI RESULTS
# ==========================

```python
total_Sales = df["Sales"].sum()
total_Profit = df["Profit"].sum()
total_Customer = df["Customer_ID"].nunique()
total_Orders = df.shape[0]
Avg_Sales = df["Sales"].mean()
Avg_Profit = df["Profit"].mean()
highest_Sales = df["Sales"].max()
lowest_Sales = df["Sales"].min()
profit_margin = (total_Profit / total_Sales) * 100

print("========== KPI Report ==========")
print(f"Total Sales: ${total_Sales:,.2f}")
print(f"Total Profit: ${total_Profit:,.2f}")
print(f"Total Customers: {total_Customer}")
print(f"Total Orders: {total_Orders}")
print(f"Average Sales: ${Avg_Sales:,.2f}")
print(f"Average Profit: ${Avg_Profit:,.2f}")
print(f"Highest Sales: ${highest_Sales:,.2f}")
print(f"Lowest Sales: ${lowest_Sales:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

Advanced EDA (Exploratory Data Analysis)

Visualisation

Top 10 Products by Revenue

```python
top_products = (
    df.groupby("Product")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products 

top_products.plot(
    kind="bar",
    figsize=(7,4),
    title="Top 10 Products by Revenue"
)

plt.show()
```

High-performing products can be featured in advertisements, email campaigns, and homepage recommendations.

```python
plt.savefig("Output/Top 10 Products by Revenue.png", dpi=300, bbox_inches="tight")
plt.show()
```

Top 10 Products by Profit

```python
top_profit_products = (
    df.groupby("Product")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_profit_products

top_profit_products.plot(
    kind="bar",
    figsize=(7,4),
    title="Total Profit by Top 10 Products"
)

plt.show()
```

Increasing the sales of profitable products has a greater impact on business growth than increasing overall sales volume.

```python
plt.savefig("Output/Total Profit by Top 10 Products.png", dpi=300, bbox_inches="tight")
plt.show()
```

Top 10 Customers by Profit

```python
top_customers_profit = (
    df.groupby("Customer_Name")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers_profit

top_customers_profit.plot(
    kind="bar",
    figsize=(7,4),
    title="Total Profit by Top 10 Customers"
)
plt.show()
```

These high-value customers should be retained through loyalty programs and exclusive offers.

```python
plt.savefig("Output/Total Profit by Top 10 Customers.png", dpi=300, bbox_inches="tight")
plt.show()
```

Average Order Value by City

Average Profit by Category

```python
category_avg_profit = (
    df.groupby("Category")["Profit"]
      .mean()
      .sort_values(ascending=False)
)

category_avg_profit

category_avg_profit.plot(
    kind="bar",
    figsize=(7,4),
    title="Average Profit by Category"
)
plt.show()
```

Payment Method Performance

```python
payment_summary = (
    df.groupby("Payment_Method")
    .agg(
        Total_Revenue=("Sales", "sum"),
        Total_Profit=("Profit", "sum"),
        Orders=("Order_ID", "count"),
        Avg_Order_Value=("Sales", "mean")    
    )
    .sort_values(by="Total_Revenue", ascending=False)
)

payment_summary

payment_summary.plot(
    kind="bar",
    figsize=(7,4),
    title="Performance by Payment Method"
)
plt.show()
```

Understanding payment preferences helps reduce cart abandonment and improve customer satisfaction.

```python
plt.savefig("Output/Performance by Payment Method.png", dpi=300, bbox_inches="tight")
plt.show()
```

Shipping Mode Performance

```python
shipping_summary = (
    df.groupby("Ship_Mode")
    .agg(
        Total_Revenue=("Sales", "sum"),
        Total_Profit=("Profit", "sum"),
        Orders=("Order_ID", "count"),
    )
    .sort_values(by="Total_Revenue", ascending=False)
)
shipping_summary

shipping_summary.plot(
    kind="bar",
    figsize=(7,4),
    title="Performance by Shipping Mode"
)
plt.show()
```

Standard Shipping generally accounts for the highest number of orders due to lower delivery costs.

```python
plt.savefig("Output/Performance by Shipping Mode.png", dpi=300, bbox_inches="tight")
plt.show()
```

City Performance Dashboard

```python
city_dashboard = (
    df.groupby("City")
    .agg(
        Revenue=("Sales", "sum"),
        Profit=("Profit", "sum"),
        Orders=("Order_ID", "count"),
        Customers=("Customer_ID", "nunique"),
    )
    .sort_values(by="Revenue", ascending=False) 
)
city_dashboard

city_dashboard.plot(    
    kind="bar",
    figsize=(7,4),
    title="City Dashboard"
)
plt.show()
```

The top-performing city contributes the highest share of total revenue, indicating strong customer demand in that region.

```python
plt.savefig("Output/City Dashboard.png", dpi=300, bbox_inches="tight")
plt.show()
```
